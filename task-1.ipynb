```python
# 8. for 100 datapoints chosen randomly plot price v/s stock chart. 
import matplotlib.pyplot as plt
import matplotlib

matplotlib.style.use('ggplot')

random = df.sample(n=100)
plt.bar(random['price'] , random['stock'])
plt.xlabel("price") 
plt.ylabel("stock")
```
